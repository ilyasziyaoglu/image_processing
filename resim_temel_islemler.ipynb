```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# resmi renkli açma
img = cv2.imread("fas1.jpg", cv2.IMREAD_COLOR)
```

```python
# resmi alfa kanalı ile birlikte açma
img = cv2.imread("fas1.jpg", cv2.IMREAD_UNCHANGED)
```
