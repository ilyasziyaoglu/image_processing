```python
import numpy as np
import cv2
```

```python
# kamera ile video çekme
cap = cv2.VideoCapture(0) # harici bir kamerada i=0 yerine i=1,2,3..vs kullanabilirsiniz
while(True):
    ret, frame = cap.read() #Çerçeveler halinde görüntü yakalar
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY) #Üzerinde işlem yapacağımız çerçeve buraya gelsin
    cv2.imshow('frame',gray) #Sonuç Çerçeveyi Görüntüleme
    
    if cv2.waitKey(1) & 0xFF == ord('q'): # q ile çıkış yapabilirsiniz
        break

#Herşey yolunda gitti ise dükkanı kapatabiliriz :)
cap.release()
cv2.destroyAllWindows()
```

```python
# dosyadan video okumak
cap1 = cv2.VideoCapture("video1.mp4")
while(cap1.isOpened()):
    ret, frame = cap1.read() #Çerçeveler halinde görüntü yakalar
    
    if ret:
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY) #Üzerinde işlem yapacağımız çerçeve buraya gelsin
        cv2.imshow('frame',gray) #Sonuç Çerçeveyi Görüntüleme
        cv2.waitKey(25)
    else:
        break

#Herşey yolunda gitti ise dükkanı kapatabiliriz :)
cap1.release()
cv2.destroyAllWindows()
```

```python
# video kaydetme

#Codec tanımlama ve VideoWriter nesnesi(object) oluşturma
cap2 = cv2.VideoCapture("video1.mp4")
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('video2.avi', fourcc, 20.0, (1280,720))
while(cap2.isOpened()):
    ret, frame = cap2.read()
    if cv2.waitKey(1) & 0xFF == ord('q'):
       break
    if ret == True:
        frame = cv2.flip(frame,0)
        out.write(frame)
        
        cv2.imshow('frame',frame)
    else:
       break

#Herşey yolunda gitti ise dükkanı kapatabiliriz :)
cap2.release()
out.release()
cv2.destroyAllWindows()
```
