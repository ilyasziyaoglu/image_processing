```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = np.zeros((512, 512, 3), np.uint8)
```

```python
cv2.line(img, (0,0), (511,511), (255, 0, 0), 5)
plt.imshow(img)
plt.show()
```

```python
cv2.rectangle(img, (100, 100), (400, 400), (0,255,0), 3)
plt.imshow(img)
plt.show()
```

```python
cv2.circle(img,(250, 250), 150, (255,255,0), 5)
plt.imshow(img)
plt.show()
```

```python
cv2.ellipse(img,(250,250), (100,50),0,180,360,255,5)
plt.imshow(img)
plt.show()
```

```python
pts = np.array([[10,5],[20,30],[70,20],[50,10]],np.int32)
cv2.polylines(img,[pts],True,(0,255,255))
plt.imshow(img)
plt.show()
```

```python
#Beyaz renkte Esogu-EEE yazdıracağız.
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img,'Esogu-EEE', (10,500), font, 4, (255,255,255),2,cv2.LINE_AA)
plt.imshow(img)
plt.show()
```
