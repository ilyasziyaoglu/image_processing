```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# dosyadan video okumak
cap1 = cv2.VideoCapture(0)
temp = cv2.imread("temp.jpg", 0)
dif = np.zeros((480, 640), np.uint8)

while(True):
    ret, frame = cap1.read() #Çerçeveler halinde görüntü yakalar
    
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY) #Üzerinde işlem yapacağımız çerçeve buraya gelsin
    
    dif = gray - temp
    cv2.imshow('frame',dif) #Sonuç Çerçeveyi Görüntüleme
    
    if cv2.waitKey(25) & 0xFF == ord('q'): # q ile çıkış yapabilirsiniz
        break

#Herşey yolunda gitti ise dükkanı kapatabiliriz :)
cap1.release()
cv2.destroyAllWindows()
```

```python
cap1 = cv2.VideoCapture(0)

ret, frame = cap1.read() #Çerçeveler halinde görüntü yakalar

gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY) #Üzerinde işlem yapacağımız çerçeve buraya gelsin

cv2.imwrite("temp.jpg", gray)

#Herşey yolunda gitti ise dükkanı kapatabiliriz :)
cap1.release()
cv2.destroyAllWindows()
```
